Plot for cognitive test

```python
import matplotlib.pyplot as plt

interfaces = ['Clean UI', 'Chaotic UI']
recalls = [22, 11] 
recall_rate = [73, 37]

plt.figure(figsize=(8, 5))
bars = plt.bar(interfaces, recall_rate, color=['blue', 'cyan'])

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 2, f'{yval}%', ha='center', fontsize=12)

plt.title('Recall Rate by Interface Type', fontsize=14)
plt.ylabel('Recall Rate (%)')
plt.ylim(0, 100)

plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

Plots for decsion times

```python
import matplotlib.pyplot as plt

platforms = ['Duolingo', 'Udacity']
times = [5, 10] 

plt.figure(figsize=(7, 5))
bars = plt.bar(platforms, times, color=['red', 'blue'])

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, f'{yval} sec', ha='center', fontsize=12)

plt.title('Average Time to First 3 Interactions', fontsize=14)
plt.ylabel('Time (seconds)')
plt.ylim(0, 12)
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

Plot for engagagement score

```python
import matplotlib.pyplot as plt
import numpy as np

users = [f"U{i+1}" for i in range(20)]
engagement_scores = np.array([
    0.88, 0.92, 0.76, 0.81, 0.73,
    0.90, 0.85, 0.78, 0.91, 0.74,
    0.87, 0.79, 0.93, 0.80, 0.89,
    0.72, 0.83, 0.75, 0.86, 0.77
]) * 50  

average_score = round(np.mean(engagement_scores), 2)

plt.figure(figsize=(10, 6))
bars = plt.bar(users, engagement_scores, color='#B3D9FF', edgecolor='gray')

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.5, f'{height:.1f}', ha='center', fontsize=10)

plt.axhline(y=average_score, color='red', linestyle='--', linewidth=2, label=f'Avg Score = {average_score:.1f}')

plt.title("Engagement Scores (Out of 50) for Instagram Users", fontsize=14, fontweight='bold')
plt.ylabel("Engagement Score (0 – 50)")
plt.ylim(0, 55)
plt.xticks(rotation=45)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```
